```python
MENU = {
    "espresso": {
        "ingredients": {"water": 50, "coffee": 18},
        "cost": 15
    },
    "latte": {
        "ingredients": {"water": 200, "milk": 150, "coffee": 24},
        "cost": 25
    },
    "cappuccino": {
        "ingredients": {"water": 250, "milk": 100, "coffee": 24},
        "cost": 30
    }
}

resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
    "money": 0
}

def is_sufficient(ingredients):
    for item in ingredients:
        if ingredients[item] > resources[item]:
            print(f"Sorry not enough {item}.")
            return False
    return True

def process_rupees():
    print("Insert coins in ₹.")
    ones = int(input("How many ₹1 coins? "))
    twos = int(input("How many ₹2 coins? "))
    fives = int(input("How many ₹5 coins? "))
    tens = int(input("How many ₹10 coins? "))
    total = ones * 1 + twos * 2 + fives * 5 + tens * 10
    return total

def transaction_ok(money, cost):
    if money >= cost:
        change = round(money - cost, 2)
        if change > 0:
            print(f"Here is ₹{change} change.")
        resources["money"] += cost
        return True
    else:
        print("Not enough money. Refunded.")
        return False

def make_drink(name, ingredients):
    for item in ingredients:
        resources[item] -= ingredients[item]
    print(f"Here is your {name}. Enjoy!")

def report():
    print(f"Water: {resources['water']}ml")
    print(f"Milk: {resources['milk']}ml")
    print(f"Coffee: {resources['coffee']}g")
    print(f"Money: ₹{resources['money']}")

def machine():
    while True:
        choice = input("What would you like? (espresso/latte/cappuccino): ").lower()
        if choice == "off":
            break
        elif choice == "report":
            report()
        elif choice in MENU:
            drink = MENU[choice]
            if is_sufficient(drink["ingredients"]):
                money = process_rupees()
                if transaction_ok(money, drink["cost"]):
                    make_drink(choice, drink["ingredients"])
        else:
            print("Invalid choice.")

machine()
```
